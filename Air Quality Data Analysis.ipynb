It focuses on visualizing average pollution levels, detecting peak pollution times, and identifying periods with 'Good' or 'Very Unhealthy' air quality. The project also examines how often the AQI was 'Moderate' in 2018 and explores weather conditions in January and July. Insights from this analysis will help understand air pollution patterns and their seasonal variations.

```python
import pandas as pd
```

```python
data = pd.read_csv(r"C:\Users\prati\Desktop\Python\Projects\air-quality-data.csv")
```

```python
data.head()
```

```python
data.info()
```

```python
data['Timestamp'] = pd.to_datetime(data['Timestamp'], format="%d-%m-%Y %H:%M")
```

```python
data.head()
```

```python
data.rename(columns={'Timestamp':'DateTime'},inplace=True)
```

```python
data.info()
```

### The Year-wise average pollution with a Line chart.

```python
data.head()
```

```python
data['Year'] = data['DateTime'].dt.year
```

```python
data
```

```python
year_avg = data.groupby('Year')['PM2.5'].mean()
```

```python
year_avg
```

```python
year_avg.plot(kind = 'line',figsize = (15,5), marker = 'o', color = 'purple');
```

### The Area plot showing the average pollution Month wise.

```python
data.head()
```

```python
data['Month'] = data['DateTime'].dt.month
```

```python
data.head()
```

```python
month_avg = data.groupby('Month')['PM2.5'].mean()
```

```python
month_avg
```

```python
month_avg.plot(kind = 'area',figsize = (15,5));
```

###  Bar graph to show the Hour-wise average pollution.

```python
data.head()
```

```python
data['Hour'] = data['DateTime'].dt.hour
```

```python
data.head()
```

```python
avg_hour = data.groupby('Hour')['PM2.5'].mean()
```

```python
avg_hour
```

```python
avg_hour.plot(kind = 'bar',figsize = (15,5) , color = 'green');
```
